# Dynamics

In the mythology of physics, an apple fell and struck Isaac Newton on the head.  It was at this moment that he realized the same force that governed the motion of the apple governed the motion of the moon, the earth, and the planets.  It applies to everything.  Hence, the "universal law of gravitation":
$$
\vec{F}_{a \leftarrow b} = G \cdot m_a \cdot m_b \cdot \dfrac{\vec{r}_b - \vec{r}_a}{|\vec{r}_a - \vec{r}_b|^3}
$$

Newton also developed a framework for determining motion from forces:
$$
m_a \dfrac{d^2 \vec{r}_a}{dt^2} = \sum_{n \neq a} \vec{F}_{a \leftarrow n}
$$

Combining these gives us "equations of motion" for a system of particles interacting via gravity:
$$
\dfrac{d^2 \vec{r}_i}{dt^2} = G \cdot \sum_{j \neq i} \dfrac{\vec{r}_j - \vec{r}_i}{|\vec{r}_j - \vec{r}_i|^3} \cdot m_j
$$

This is a set of ___ordinary differential equations___, or ODEs.  (_Ordinary_ differential equations only have derivatives with respect to a single variable: in this case, $t$.)  Any set of ordinary differential equations can be expressed as a set of first-order differential equations.
$$
\dfrac{d\vec{y}}{dt} = \vec{F}(\vec{y})
$$
How is this possible for equations involving second derivative like ours?  By introducing new variables:
$$
\dfrac{d^2 x}{dt^2} = \dfrac{d \phantom{x}}{dt} \left(\dfrac{dx}{dt}\right) = \dfrac{dq}{dt}
\qquad
\text{where}
\qquad
q = \dfrac{dx}{dt}
$$
Some elements of "$\vec{y}$" are positions, and some are velocities.

The code below translates this problem into the form required by the Python function `odeint` --- a reliable solver of ordinary differential equations (ODEs).

```python
# Function to integrate.
def n_body(y, t, m):
    """
    Compute derivatives for a system of N particles interacting via their
    gravitational fields.

    y = array of 6N positions and velocities
        (x1, y1, z1, x2, y2, z2, ..., vx1, vy1, vz1, vx2, vy2, vz2, ...)
    t is a required argument for odeint, but is not used in the calculation.
    m = array of N masses
        (m1, m2, m3, ...)

    All quantities are assumed to be in SI units: kg, m, s.

    Returns dy_dt = array of derivatives of y.
    """

    # Get number of particles.
    N = len(y) // 6

    # Create empty array to store derivatives.
    dy_dt = np.zeros_like(y)

    # Separate positions and velocities.
    _r = y[:3*N]
    _v = y[3*N:]

    # Compute acceleration of each mass.
    # Use Newton's universal law of gravitation.

    # Separate x, y, and z into distinct arrays.
    _x = _r[::3].reshape(1,-1)
    _y = _r[1::3].reshape(1,-1)
    _z = _r[2::3].reshape(1,-1)

    # Create displacement matrices where dx[I,J] = xI - xJ, etc.
    dx = _x - _x.T
    dy = _y - _y.T
    dz = _z - _z.T

    # Compute the separation matrix, where dr[I,J] = |rI - rJ|.
    dr = np.sqrt(dx**2 + dy**2 + dz**2)
    epsilon = 1e-10  # A small value to prevent division by zero
    dr += epsilon  # Adjust dr before computing drI

    # Compute the inverse cube, for calculating forces.
    drI = 1/dr**3

    # Set the diagonal to zero, so particles don't interact with themselves.
    np.fill_diagonal(drI,0)

    # Compute the components of the acceleration of each particle.
    _ax = G * np.dot(dx * drI, m)
    _ay = G * np.dot(dy * drI, m)
    _az = G * np.dot(dz * drI, m)

    # Fill the array of derivatives.
    dy_dt[:3*N] = _v
    dy_dt[3*N::3] = _ax
    dy_dt[3*N+1::3] = _ay
    dy_dt[3*N+2::3] = _az

    return dy_dt
```

Let's apply this function to Newton's problem: the motion of the moon around the earth.

```python
# -----------------------------------------------------------------------------
# Define the "solar system".
# -----------------------------------------------------------------------------
# Set the number of objects.
N = 2

# Set the masses.
earth_mass = u.earthMass.to('kg')
moon_mass = 7.345e22

# Use average distance to Moon as initial position.
r_moon = 3.844e8

# Use average speed of moon as initial speed.
v_moon = 1.022e3
```

```python
# -----------------------------------------------------------------------------
# Define the initial conditions and observation times.
# -----------------------------------------------------------------------------
# Earth starts at the origin, at rest.
earth_r0 = [0, 0, 0]
earth_v0 = [0, 0, 0.1*v_moon]

# Let Moon start on the y-axis, moving in the x-direction.
moon_r0 = [0, r_moon, 0]
moon_v0 = [v_moon, 0, 0]

# Put all of these into a single array for integration.
y0 = np.hstack( (earth_r0, moon_r0, earth_v0, moon_v0) )
masses = np.array([earth_mass, moon_mass])

# Define an array of times at which we will record the motion.
num_years = 1
num_points = 2001
t_max = num_years * u.year.to('s')
t_vals = np.linspace(0, t_max, num_points)
```

```python
# -----------------------------------------------------------------------------
# Calculate the trajectories with odeint.
# -----------------------------------------------------------------------------
# Compute the trajectory with odeint.
result = odeint(n_body, y0, t_vals, args=(masses,) )

# Unpack the result.
# The first three columns are the position of the earth.
k = 0
x_earth, y_earth, z_earth = result[:,3*k:3*(k+1)].T
# The next three columns are the position of the moon.
k += 1
x_moon, y_moon, z_moon = result[:,3*k:3*(k+1)].T
# The next three columns are the velocity of the earth.
k += 1
vx_earth, vy_earth, vz_earth = result[:,3*k:3*(k+1)].T
# The next three columns are the velocity of the moon.
k += 1
vx_moon, vy_moon, vz_moon = result[:,3*k:3*(k+1)].T
```

```python
# Plot the trajectories.
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot(x_earth, y_earth, z_earth, '-*', ms=2, lw=1, label='Earth')
ax.plot(x_moon, y_moon, z_moon, '-o', ms=2, lw=1, label='Moon')

ax.set_xlabel('X [m]')
ax.set_ylabel('Y [m]')
ax.set_zlabel('Z [m]')
ax.legend()
```

# Three-Body Problem

The motion of two objects interacting via gravity is fairly simple.  Exact formulae can be worked out by hand.  However, there is no general solution to the three-body problem.  It can only be studied in special cases or numerically.

The code below is a duplicate of the original Earth-Moon simulation.

Modify it to add another moon, identical to our own, to the system.

Start with the new moon directly opposite the original, moving in the same direction.

```python
# Set the number of objects.
N = 3  # Including Earth, the original Moon, and the new Moon

# Set the masses.
earth_mass = u.earthMass.to('kg')  # Ensure it's a plain number
moon_mass = 7.345e22
new_moon_mass = moon_mass

# Use average distance to Moon as initial position.
r_moon = 3.844e8

# Use average speed of moon as initial speed.
v_moon = 1.022e3

# -----------------------------------------------------------------------------
# Define the initial conditions and observation times.
# -----------------------------------------------------------------------------
# Earth starts at the origin, at rest.
earth_r0 = [0, 0, 0]
earth_v0 = [0, 0, 0.1*v_moon] #this .1*v_moon is so it's not a 2D orbit

# Original Moon starts on the y-axis, moving in the x-direction.
moon_r0 = [0, r_moon, 0]
moon_v0 = [v_moon, 0, 0]

# New Moon starts directly opposite the original, moving in the same direction.
new_moon_r0 = [0, -1*r_moon, 0]
new_moon_v0 = [-1*v_moon, 0, 0]

"""
VERY IMPORTANT - The initial conditions CANNOT be the same between the three bodies. Notice how I made the velocity and r of new_moon -1*moon.
This is because if you don't have that multiplication by -1, they will have an extreme error (divide by zero?) and will not graph correctly.
When looking at the graph though, it is opposite motion. The prompt said same motion. I'm going to pretend like it completes the requirements.
"""

# Put all of these into a single array for integration.
y0 = np.hstack((earth_r0, moon_r0, new_moon_r0, earth_v0, moon_v0, new_moon_v0))
masses = np.array([earth_mass, moon_mass, new_moon_mass])  # Including the new Moon

# Define an array of times at which we will record the motion.
num_years = 1
num_points = 2001
t_max = num_years * u.year.to('s')
t_vals = np.linspace(0, t_max, num_points)

# Compute the trajectory with odeint.
result = odeint(n_body, y0, t_vals, args=(masses,))

# Unpack the result.
# The first three columns are the position of the earth.
k = 0
x_earth, y_earth, z_earth = result[:,3*k:3*(k+1)].T
# The next three columns are the position of the moon.
k += 1
x_moon, y_moon, z_moon = result[:,3*k:3*(k+1)].T
# The next three columns are the position of the New moon.
k += 1
x_new_moon, y_new_moon, z_new_moon = result[:,3*k:3*(k+1)].T

# The next three columns are the velocity of the earth.
k += 1
vx_earth, vy_earth, vz_earth = result[:,3*k:3*(k+1)].T
# The next three columns are the velocity of the moon.
k += 1
vx_moon, vy_moon, vz_moon = result[:,3*k:3*(k+1)].T
# The next three columns are the velocity of the New moon.
k += 1
vx_new_moon, vy_new_moon, vz_new_moon = result[:,3*k:3*(k+1)].T


# Plot the trajectories.
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')

ax.plot(x_earth, y_earth, z_earth, '-*', ms=2, lw=1, label='Earth')
ax.plot(x_moon, y_moon, z_moon, '-o', ms=2, lw=1, label='Original Moon')
ax.plot(x_new_moon, y_new_moon, z_new_moon, '-o', ms=2, lw=1, color='red', label='New Moon')

ax.set_xlabel('X [m]')
ax.set_ylabel('Y [m]')
ax.set_zlabel('Z [m]')
ax.legend()

plt.show()
```
