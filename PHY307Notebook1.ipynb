```python
# Let's get some astronomical data from AstroPy.
earth_radius = 1 * u.earthRad
earth_mass = 1 * u.earthMass
earth_orbit = 1 * u.astronomical_unit
earth_period = 1 * u.year

print(earth_radius)
print(earth_mass)
print(earth_orbit)
print(earth_period)
```

The commands above define ___objects___ that have data and methods (functions) attached to them.  They are not simple numbers or functions.  We can add or subtract quantity with the same dimensions (length, mass, time, etc.).

```python
#earth_radius + earth_mass
```

We can, however, form new quantities through multiplication and division.

```python
earth_rotation_speed = 2 * pi * earth_radius / (1 * u.day)
```

What is the speed of the earth?

```python
earth_rotation_speed
```

True, but not necessarily useful.  We can convert to any SI units, though.

```python
earth_rotation_speed.to('m/s')
```

# Coordinates

Unit conversions are nice, but you don't need Python to do the calculations.  However, coordinate transformations are tricky.  AstroPy is quite helpful.

We'll plot the local coordinates of the sun and moon, following the example here:

<https://docs.astropy.org/en/stable/generated/examples/coordinates/plot_obs-planning.html>

First, let's find the $(x,y,z)$ coordinates of Oregon Tech, relative to the center of the earth.

```python
import astropy.coordinates as coord
from astropy.coordinates import AltAz, SkyCoord, EarthLocation
from astropy.time import Time
```

```python
# Define the location of Oregon Tech.
lattitude = 42.256
longitude = -121.785
elevation = 1300 * u.m
```

```python
# Get the (x,y,z) coordinates.
klamath_falls = EarthLocation.from_geodetic(longitude, lattitude, elevation)
# Compare with the North Pole.
north_pole = EarthLocation.from_geodetic(0,90,0)
```

```python
print(klamath_falls)
print(north_pole)
```

Timing is important in astronomy.  The `Time` object we imported above is very helpful.  Let's set our local clock.  Pacific Standard Time is -8 hours from Coordinated Universal Time (UTC).

```python
utc_offset = -8 * u.hour  # Pacific Standard Time
midnight = Time('2024-1-18 00:00:00') - utc_offset
```

Let's see where Sirius will be at midnight tonight ...

```python
# Get the location of Sirius on the celestial sphere.
sirius_celestial = SkyCoord.from_name('Sirius')

# Get the local coordinates of Sirius at midnight.
sirius_local = sirius_celestial.transform_to(AltAz(obstime=midnight, location=klamath_falls))

# Use Polaris for comparison.
polaris_celestial = SkyCoord.from_name('Polaris')
polaris_local = polaris_celestial.transform_to(AltAz(obstime=midnight, location=klamath_falls))

print("Sirius ...")
print(sirius_celestial)
print(sirius_local)

print("Polaris ...")
print(polaris_celestial)
print(polaris_local)
```

Next, we will create an array of times for calculating positions.

```python
# Create an array of times around midnight.
delta = np.linspace(-12,12,144) * u.hour
time_frame = AltAz(obstime = midnight + delta, location = klamath_falls)
# print(time_frame)
```

Now, let's find the location of the sun (to see if it is dark), the moon, and Sirius during our 24-hour window.

```python
# Use the celestial coordinates of Sirius and the local time
# to find its location in the sky.
sirius_today = sirius_celestial.transform_to(time_frame)

# Use a special function "get_body" to find the location of the sun and moon.
sun_today = coord.get_body( "sun", midnight + delta).transform_to(time_frame)
moon_today = coord.get_body("moon", midnight + delta).transform_to(time_frame)
```

Now let's plot the positions of the sun, moon, and Sirius.

```python
# Import plotting libraries and configure settings.
import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style, quantity_support

plt.style.use(astropy_mpl_style)
quantity_support()
```

```python
# Plot the location of Sirius above the horizon.
# Use color to indicate the azimuth.
plt.plot(delta, sun_today.alt, color='r', label='Sun')
plt.plot(delta, moon_today.alt, color='y', ls='--', label='Moon')
plt.scatter(delta, sirius_today.alt, c=sirius_today.az.value,
            label='Sirius', lw=0, s=8, cmap='viridis')
plt.fill_between(delta, -90*u.deg, 90*u.deg,
                 sun_today.alt < -0*u.deg, color='0.5', zorder=0)
plt.fill_between(delta, -90*u.deg, 90*u.deg,
                 sun_today.alt < -18*u.deg, color='k', zorder=0)
plt.colorbar().set_label('Azimuth [deg]')
plt.legend(loc='upper left')
plt.xlim(-12*u.hour, 12*u.hour)
plt.xticks((np.arange(13)*2-12)*u.hour)
plt.ylim(-90*u.deg, 90*u.deg)
plt.xlabel('Hours from PST Midnight')
plt.ylabel('Altitude [deg]')
plt.show()

# This plot was adapted with little modification from
# https://docs.astropy.org/en/stable/generated/examples/coordinates/plot_obs-planning.html
```

## Eclipse

There _will_ be an eclipse on April 8, 2024, in El Salto, Mexico.  Make a plot that shows when this will occur.

```python
# Create an array of times around midnight.
delta = np.linspace(-12,12,144) * u.hour
#print(delta)

utcoffset = -6*u.hour  # Central Standard Time Offset

time = Time('2024-4-08 13:09:29') - utcoffset
#print(time)

# Define the location of El Salto.
lattitude_ES = 20.52
longitude_ES = -103.18
elevation_ES = 979 * u.m

# Get the (x,y,z) coordinates.
El_Salto = EarthLocation.from_geodetic(longitude_ES, lattitude_ES, elevation_ES)

eclipseday = AltAz(obstime = time + delta, location = El_Salto)

# Use a special function "get_body" to find the location of the sun and moon.
sun_eclipse = coord.get_body( "sun", time + delta).transform_to(eclipseday)
moon_eclipse = coord.get_body("moon", time + delta).transform_to(eclipseday)

# Import plotting libraries and configure settings.
import matplotlib.pyplot as plt
from astropy.visualization import astropy_mpl_style, quantity_support

plt.style.use(astropy_mpl_style)
quantity_support()


plt.plot(delta, sun_eclipse.alt, color='r', label='Sun')

plt.scatter(delta, moon_eclipse.alt, color='y',
            label='Moon', lw=0, s=8)

plt.fill_between(delta, -90*u.deg, 90*u.deg,
                 sun_eclipse.alt < -0*u.deg, color='0.5', zorder=0)

plt.fill_between(delta, -90*u.deg, 90*u.deg,
                 sun_eclipse.alt < -18*u.deg, color='k', zorder=0)

plt.legend(loc='upper left')
plt.xlim(-12*u.hour, 12*u.hour)
plt.xticks((np.arange(13)*2-12)*u.hour)
plt.ylim(-90*u.deg, 90*u.deg)
plt.xlabel('Hours from Eclipse')
plt.ylabel('Altitude [deg]')
plt.show()
```

Calculate how close the centers of the sun and moon will pass (in degrees) at the time of closest approach during the eclipse.

```python
# Create an array of times around midnight.
delta = np.linspace(-12, 12, 144) * u.hour

utcoffset = -6 * u.hour # Central Standard Time Offset

time = Time('2024-4-08 13:09:29') - utcoffset

# Define the location of El Salto.
latitude_ES = 20.52
longitude_ES = -103.18
elevation_ES = 979 * u.m

# Get the (x,y,z) coordinates.
El_Salto = EarthLocation.from_geodetic(longitude_ES, latitude_ES, elevation_ES)

eclipseday = AltAz(obstime=time + delta, location=El_Salto)

# Use a special function "get_body" to find the location of the sun and moon.
sun_eclipse = coord.get_body( "sun", time + delta).transform_to(eclipseday)
moon_eclipse = coord.get_body("moon", time + delta).transform_to(eclipseday)

# Use Astropys seperation
separation = sun_eclipse.separation(moon_eclipse)

min_separation = np.min(separation)
closest_time = time + delta[np.argmin(separation)]

print(f"Closest approach occurs at {closest_time.iso} with a separation of {min_separation}.")
```
