```python
import matplotlib.pyplot as plt
import numpy as np

# Data for HD 189733b and its star
orbital_distance_au = 0.031  # in astronomical units
orbital_period_days = 2.2  # in Earth days

# Labels for the graph
labels = ['HD 189733b Orbit']
distance_values = [orbital_distance_au]
period_values = [orbital_period_days]

# Creating the graph
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Exoplanet')
ax1.set_ylabel('Orbital Distance (AU)', color=color)
ax1.bar(labels, distance_values, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Orbital Period (Days)', color=color)  
ax2.plot(labels, period_values, color=color, marker='o', linestyle='None')
ax2.tick_params(axis='y', labelcolor=color)

# Otherwise the right y-label is slightly clipped
fig.tight_layout()  
plt.title('HD 189733b Orbital Distance and Period')
plt.show()
```
